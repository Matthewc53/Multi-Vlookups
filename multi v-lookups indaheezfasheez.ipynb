```python
#read in edge extract and source file

df = pd.read_excel('Pandas Custom\Standard Office Systems - Rerate - Edge Format - Item Level with Costs.xlsx')

df1 = pd.read_excel('Pandas Custom\Rating_Report_2024_05_09_13_26_29.xls')
```

```python
#create slice for each carrier

dfx = df1[df1['Carrier SCAC']=='AACT']

dfy = df1[df1['Carrier SCAC']=='FWDA']

dfz= df1[df1['Carrier SCAC']=='FXNL']

dfa = df1[df1['Carrier SCAC']=='SEFL']

dfb = df1[df1['Carrier SCAC']=='TFIN']

dfc = df1[df1['Carrier SCAC']=='RLCA']

dfd = df1[df1['Carrier SCAC']=='CTII']

dfe = df1[df1['Carrier SCAC']=='DYLT']

```

```python
#perform right join on Primary Reference to combine source and edge output data 

dfx1 = dfx.merge(df, how='right',on='Primary Reference',indicator=True)

dfy1 = dfy.merge(df, how='right',on='Primary Reference',indicator=True)

dfz1 = dfz.merge(df, how='right',on='Primary Reference',indicator=True)

dfa1 = dfa.merge(df, how='right',on='Primary Reference',indicator=True)

dfb1 = dfb.merge(df, how='right',on='Primary Reference',indicator=True)

dfc1 = dfc.merge(df, how='right',on='Primary Reference',indicator=True)

dfd1 = dfd.merge(df, how='right',on='Primary Reference',indicator=True)

dfe1 = dfe.merge(df, how='right',on='Primary Reference',indicator=True)
```

```python
#add in new LH column as per request for each carrier

dfx1['LH'] = dfx1['Normalized Total']-dfx1['Fuel Total']-dfx1['Accessorial Total']

dfy1['LH'] = dfy1['Normalized Total']-dfy1['Fuel Total']-dfy1['Accessorial Total']

dfz1['LH'] = dfz1['Normalized Total']-dfz1['Fuel Total']-dfz1['Accessorial Total']

dfa1['LH'] = dfa1['Normalized Total']-dfa1['Fuel Total']-dfa1['Accessorial Total']

dfb1['LH'] = dfb1['Normalized Total']-dfb1['Fuel Total']-dfb1['Accessorial Total']

dfc1['LH'] = dfc1['Normalized Total']-dfc1['Fuel Total']-dfc1['Accessorial Total']

dfd1['LH'] = dfd1['Normalized Total']-dfd1['Fuel Total']-dfd1['Accessorial Total']

dfe1['LH'] = dfe1['Normalized Total']-dfe1['Fuel Total']-dfe1['Accessorial Total']
```

```python
#select only wanted columns for each carrier selection

dfx1 = dfx1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfy1 = dfy1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfz1 = dfz1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfa1 = dfa1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfb1 = dfb1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfc1 = dfc1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfd1 = dfd1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]

dfe1 = dfe1[['Normalized Total', 'Carrier SCAC', 'Primary Reference','LH', 'Fuel Total', 'Accessorial Total']]
```

```python
new = pd.concat([dfx1,dfy1, dfz1,dfa1,dfb1,dfc1,dfd1,dfe1], axis=1)
```
